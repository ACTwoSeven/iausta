```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

#Ejercicio 1
Graficar variables de temperatura

```python
temp = ctrl.Antecedent(np.arange(80, 110, 1), 'Temperatura')
temp['Alta']=fuzz.trapmf(temp.universe, [80, 80, 90, 100])
temp['MuyAlta']=fuzz.trapmf(temp.universe, [90, 100, 110,110])
temp.view()
```
